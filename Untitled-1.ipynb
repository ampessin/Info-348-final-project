```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
from collections import Counter

# Genius API setup
BASE_URL = "https://api.genius.com"
TOKEN = "j27rcomZ8sLHGxJoUyVA8tfJ2poJjaiLRzXr6TqkhPw7yBG5D7-Xx8sjWwF5lxA4"
HEADERS = {"Authorization": f"Bearer {TOKEN}"}

# Words to track
FILLER = ["like", "yeah", "uh", "um", "you know", "huh", "ayy", "yo", "ooh", "hey"]
SWEARS = ["fuck", "shit", "bitch", "ass", "damn", "hoe"]

# Simple text cleaning
def clean_text(text):
    text = text.lower()
    text = re.sub(r"[^\w\s]", "", text)  # remove punctuation
    for f in FILLER:
        text = re.sub(r"\b" + re.escape(f) + r"\b", "", text)
    return text

# Count words, filler words, swears
def count_words(text):
    words = text.split()
    word_counts = Counter(words)
    filler_count = sum(word_counts[f] for f in FILLER if f in word_counts)
    swear_count = sum(word_counts[s] for s in SWEARS if s in word_counts)
    return word_counts.most_common(10), filler_count, swear_count

# Genius API search
def search_artist(artist_name):
    url = BASE_URL + "/search"
    params = {"q": artist_name}
    response = requests.get(url, params=params, headers=HEADERS)
    return response.json()

# Get song lyrics from URL
def extract_lyrics_from_url(url):
    page = requests.get(url)
    html = BeautifulSoup(page.text, "html.parser")
    lyrics_div = html.find("div", class_="lyrics") or html.find_all("div", {"data-lyrics-container": "true"})
    if lyrics_div:
        if isinstance(lyrics_div, list):
            return " ".join([d.get_text(separator=" ") for d in lyrics_div]).strip()
        return lyrics_div.get_text(separator=" ").strip()
    return ""

# Define artists by genre
GENRES = {
    "rap": ["Drake", "Kendrick Lamar", "Travis Scott"],
    "pop": ["Taylor Swift", "Ariana Grande", "Olivia Rodrigo"],
    "country": ["Luke Combs", "Morgan Wallen", "Kane Brown"]
}

all_songs = []

# Collect data
for genre, artists in GENRES.items():
    for artist in artists:
        search_result = search_artist(artist)
        hits = search_result["response"]["hits"][:3]  # top 3 songs per artist
        for hit in hits:
            song_title = hit["result"]["title"]
            song_url = hit["result"]["url"]
            lyrics = extract_lyrics_from_url(song_url)
            if not lyrics:
                continue  # skip songs without lyrics
            cleaned = clean_text(lyrics)
            top_words, filler_count, swear_count = count_words(cleaned)

            all_songs.append({
                "genre": genre,
                "artist": artist,
                "song_title": song_title,
                "lyrics": lyrics,
                "clean_lyrics": cleaned,
                "top_words": top_words,
                "filler_count": filler_count,
                "swear_count": swear_count
            })

# Save to CSV
df = pd.DataFrame(all_songs)
df.to_csv("simplified_lyrics_dataset.csv", index=False)
print("Dataset saved! Shape:", df.shape)
```

```python
import matplotlib.pyplot as plt

# Filler words by genre
filler_by_genre = df.groupby("genre")["filler_count"].sum()
filler_by_genre.plot(kind="bar", color="skyblue")
plt.title("Total Filler Words by Genre")
plt.ylabel("Count")
plt.show()
```
